```python
#importing libs (1)
import pandas as pd

import numpy as np

import seaborn as sns
```

```python
#load iris ds (2)
iris = sns.load_dataset('iris')
iris.head()
```

```python
#drop species column (3)
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
#drop species alternate (3.5)
iris.drop('species', axis=1, inplace=True)
iris.head()
```

```python
#add sum columns (4)
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
#create new data frame (5)
summary = pd.DataFrame({
'mean': iris.mean(),
'std': iris.std(),
'min': iris.min(),
'max': iris.max()
})
print(summary)
```
